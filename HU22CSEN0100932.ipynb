```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic(x, a=-1, b=2, c=5):
    return a * x**2 + b * x + c

# Define the stages of the Waterfall Model
stages = ["Requirements", "Design", "Implementation", "Verification", "Maintenance"]
x_positions = np.linspace(-2, 2, len(stages))  # Equally spaced x positions for stages
y_positions = quadratic(x_positions)  # Compute y positions using the quadratic function

# Plot the quadratic curve for the waterfall model
x_curve = np.linspace(-2.5, 2.5, 500)  # Smoother curve for the quadratic function
y_curve = quadratic(x_curve)

# Create the figure for plotting
plt.figure(figsize=(10, 6))

# Plot the quadratic curve
plt.plot(x_curve, y_curve, color="darkorange", linestyle="-", label="Quadratic Curve", linewidth=2)

# Add markers and connect them with vertical lines to create the waterfall effect
for i in range(len(stages)):
    if i > 0:
        # Draw a vertical drop to simulate a waterfall
        plt.plot(
            [x_positions[i-1], x_positions[i]],  # Horizontal segment between the previous and current stage
            [y_positions[i-1], y_positions[i-1]],  # Horizontal line at the previous stage's y-value
            color="darkblue", linewidth=2
        )
        plt.plot(
            [x_positions[i], x_positions[i]],  # Vertical drop at the current stage
            [y_positions[i-1], y_positions[i]],  # Drop from previous stage to current stage's y-value
            color="darkblue", linewidth=2
        )

    # Place markers and labels for each stage
    plt.scatter(x_positions[i], y_positions[i], color="red", zorder=5, s=100, edgecolors="black")  # Larger red markers
    plt.text(x_positions[i], y_positions[i] + 0.4, stages[i], ha="center", fontsize=12, fontweight="bold")

# Style the plot
plt.title("Waterfall Model Representation Using Quadratic Curve", fontsize=16, fontweight="bold")
plt.xlabel("Stages", fontsize=14)
plt.ylabel("Value (Model Output)", fontsize=14)
plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
plt.axvline(0, color="black", linewidth=0.7, linestyle="--")
plt.grid(color="gray", linestyle="--", linewidth=0.7, alpha=0.7)

# Add the legend
plt.legend(loc="upper left", fontsize=12)

# Display the plot
plt.show()
```
